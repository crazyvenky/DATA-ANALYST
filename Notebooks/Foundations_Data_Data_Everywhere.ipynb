# SIX STEP PROCESS
1. <b>Ask:</b> business challenge, objective, or question

2. <b>Prepare:</b> data generation, collection, storage, and data management

3. <b>Process:</b> data cleaning and data integrity

4. <b>Analyze:</b> data exploration, visualization, and analysis

5. <b>Share:</b> communicating and interpreting results 

6. <b>Act:</b>  putting  insights to work to solve the problem

    <div class="dropdown-content">
        <p>❓First up, the analysts needed to define what the project would look like and what would qualify as a successful result. So, to determine these things, they asked effective questions and collaborated with leaders and managers who were interested in the outcome of their people analysis. These were the kinds of questions they asked:</p>
<p> 🍧  What do you think new employees need to learn to be successful in their first year on the job?</p>
<p> 🍧  Have you gathered data from new employees before? If so, may we have access to the historical data?</p>
<p> 🍧  Do you believe managers with higher retention rates offer new employees something extra or unique?</p>
<p> 🍧  What do you suspect is a leading cause of dissatisfaction among new employees?</p>
<p> 🍧  By what percentage would you like employee retention to increase in the next fiscal year?</p>

</div>
</details>

<details>
    <summary><img src="../Images/prepare.png" alt="Image Description"></summary>
    <hr>
    <div class="dropdown-content">
        <p>🚀 It all started with solid preparation. The group built a timeline of three months and decided how they wanted to relay their progress to interested parties. Also during this step, the analysts identified what data they needed to achieve the successful result they identified in the previous step - in this case, the analysts chose to gather the data from an online survey of new employees. These were the things they did to prepare:</p>
<p>💡 They developed specific questions to ask about employee satisfaction with different business processes, such as hiring and onboarding, and their overall compensation.</p>
<p>💡 They established rules for who would have access to the data collected - in this case, anyone outside the group wouldn't have access to the raw data, but could view summarized or aggregated data. For example, an individual's compensation wouldn't be available, but salary ranges for groups of individuals would be viewable.</p>
<p>💡 They finalized what specific information would be gathered, and how best to present the data visually. The analysts brainstormed possible project- and data-related issues and how to avoid them.</p>

</div>
</details>

<details>
    <summary><img src="../Images/process.png" alt="Image Description"></summary>
    <hr>
    <div class="dropdown-content">
        <p>🌟 The group sent the survey out. Great analysts know how to respect both their data and the people who provide it. Since employees provided the data, it was important to make sure all employees gave their consent to participate. The data analysts also made sure employees understood how their data would be collected, stored, managed, and protected. Collecting and using data ethically is one of the responsibilities of data analysts. In order to maintain confidentiality and protect and store the data effectively, these were the steps they took:</p>
<p>🛡️ They restricted access to the data to a limited number of analysts.</p>
<p>🧹 They cleaned the data to make sure it was complete, correct, and relevant. Certain data was aggregated and summarized without revealing individual responses.</p>
<p>🔒 They uploaded raw data to an internal data warehouse for an additional layer of security.</p>

</div>
</details>

<details>
    <summary><img src="../Images/analyze.png" alt="Image Description"></summary>
    <hr>
    <div class="dropdown-content">
        <p>🔍 Then, the analysts did what they do best: analyze! From the completed surveys, the data analysts discovered that an employee’s experience with certain processes was a key indicator of overall job satisfaction. These were their findings:</p>
<p>🔎 Employees who experienced a long and complicated hiring process were most likely to leave the company.</p>
<p>🔬 Employees who experienced an efficient and transparent evaluation and feedback process were most likely to remain with the company.</p>
<p>📊 The group knew it was important to document exactly what they found in the analysis, no matter what the results. To do otherwise would diminish trust in the survey process and reduce their ability to collect truthful data from employees in the future.</p>

</div>
</details>

<details>
    <summary><img src="../Images/share.png" alt="Image Description"></summary>
    <hr>
    <div class="dropdown-content">
        <p>📤 Just as they made sure the data was carefully protected, the analysts were also careful sharing the report. This is how they shared their findings:</p>
<p>📊 They shared the report with managers who met or exceeded the minimum number of direct reports with submitted responses to the survey.</p>
<p>👥 They presented the results to the managers to make sure they had the full picture.</p>
<p>🗣️ They asked the managers to personally deliver the results to their teams.</p>
<p>🔄 This process gave managers an opportunity to communicate the results with the right context. As a result, they could have productive team conversations about next steps to improve employee engagement.</p>

        
</div>
</details>

<details>
    <summary><img src="../Images/act.png" alt="Image Description"></summary>

    <div class="dropdown-content">
        <p>🔄 The last stage of the process for the team of analysts was to work with leaders within their company and decide how best to implement changes and take actions based on the findings. These were their recommendations:</p>
<p>📋 Standardize the hiring and evaluation process for employees based on the most efficient and transparent practices.</p>
<p>🔄 Conduct the same survey annually and compare results with those from the previous year.</p>
<p>📊 A year later, the same survey was distributed to employees. Analysts anticipated that a comparison between the two sets of results would indicate that the action plan worked. Turns out, the changes improved the retention rate for new employees and the actions taken by leaders were successful!</p>

#### Note: Iteration during the data analysis process is not linear

## Why gut instinct can be a problem?
 Data analysts and detectives share a similar approach to problem-solving, both relying on evidence and facts to make decisions. Data-driven decision-making is essential for analysts, but gut instinct can also play a role in identifying patterns and connections. Balancing data and gut instinct is crucial for making informed decisions, and the right mix depends on the project's goals and time constraints.
